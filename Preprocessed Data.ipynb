```python
import pandas as pd
import numpy as np
data1 = pd.read_csv("C:/Users/Ivy/Desktop/data scientist/Python/Week 4/sales.csv")
data1['Price'] = data1['Price'].str.replace(',', '').astype(np.float64)
data1.insert(12, "standardized_price", 0) #insert column

#Standardized Price
data1.standardized_price = (data1.Price - np.mean(data1.Price))/(np.std(data1.Price))
print(data1[['Price','standardized_price']])

#check:
mean_std = data1['standardized_price'].mean()
stdv = data1['standardized_price'].std()
print("mean:",mean_std)
print("standard deviation:",stdv)

#store to new sales.csv
data1.to_csv("C:/Users/Ivy/Desktop/data scientist/Python/Week 4/Enopia_sales.csv",index = False)
```

```python
#test for skewness
median = data1['Price'].median()
mean = data1['Price'].mean()
print("median:",median)
print("mean:",mean)

if  mean < median:
    print ("negatively skewed")
else:
    print ("positively skewed")
```

##### Two methods can be used in determining outliers. For skewed distribution, it is better to use the IQR method. For comparison purposes, results using the two methods were presented below.

```python
#First method: using standard deviation(3-Sigma Limits) to remove outliers for the Price column
#This method will be applied to both price column and the standardized price column
import numpy as np

mean = np.mean(data1.Price)
std = np.std(data1.Price)
print("mean:",mean)
print("standard deviation:",std)


ll = max(0,mean - 3*std) #lower limit
ul =mean + 3*std #upper limit
print("lower limit:",ll)
print("upper limit:",ul)

cleandata1 = data1[data1.Price >=ll]
cleandata2 = data1[data1.Price <=ul]

cleandata2.describe()
```

##### Note:After cleaning, mean decreases to a significant value. Outliers (count) have 999-981 = 18 values

```python
#First method: using standard deviation for the standardized_price column (3-Sigma Limits)

import numpy as np

mean = np.mean(data1.standardized_price)
std = np.std(data1.standardized_price)
print("mean:",mean)
print("standard deviation:",std)

ll_std = max(0,mean - 3*std) #lower limit
ul_std =mean + 3*std #upper limit
print("lower limit:",ll_std)
print("upper limit:",ul_std)

cleandata1_std = data1[data1.standardized_price >=ll_std]
cleandata2_std = data1[data1.standardized_price <=ul_std]
cleandata2_std.describe()
```

```python
#Second method: using Inter-Quartile range
IQR = np.percentile(data1.Price,75)-np.percentile(data1.Price,25)
print(IQR)

ul2=np.percentile(data1.Price,75) + 1.5*IQR
ll2=max(0,np.percentile(data1.Price,25) - 1.5*IQR)

print(ul2)
print(ll2)
```

##### Note: np.percentile treats nan's as very high/infinite numbers. So the high percentiles will be in the range where you will end up with a nan. To calculate the percentile without the nan, np.nanpercentile is used

```python

IQR = np.nanpercentile(data1.Price,75)-np.nanpercentile(data1.Price,25)
print(IQR)

ul2=np.nanpercentile(data1.Price,75) + 1.5*IQR
ll2=max(0,np.nanpercentile(data1.Price,25) - 1.5*IQR)

print(ul2)
print(ll2)

cleandata3=data1[data1.Price >= ll2]
cleandata4=data1[data1.Price <= ul2]
cleandata4.describe()
```

##### Observation: IQR removes higher number of outliers than the standard deviation. Check that cleaned data count under the price column is 981 in the first method (both using the standardized price column and the original price data) while 841 in IQR method. 
#### *** Since we have positively skewed price distribution, it is better to use the results from the IQR method***.

```python
data2 = pd.read_csv("C:/Users/Ivy/Desktop/data scientist/Python/Week 4/Handling_Missing_Values.csv")
print(data2)

print("---------------------------------------------------------")
print("missing values per column:")
data2.isnull().sum() #determining missing values
```

```python
drop = data2.dropna()
print(drop)

print("---------------------------------------------------------")
print("missing values per column after dropping:")
drop.isnull().sum() #determining missing values after dropping 6 entries
```

##### Note: Since dropped entries consist of 30% of 20 entries, it is not feasible to do this method as well as deleting and removing data. Rather, preserving these data is a better option. For the categorical columns such as Sex and Children, missing data should be filled with "No Data" since these are personal information. It is more logical to store personal information as part of data protection for clients. However, the missing values in the income column may be replaced with estimates for statistical analysis since we only have 20 entries.

```python

#preserving data
data2[['Sex','Children']]= data2[['Sex','Children']].fillna("No Data")#fill NaN vlues
print(data2)


print("---------------------------------------------------------")
print("missing values per column after filling with 'No Data':")
data2.isnull().sum()
```

```python
#test for skewness
mn = data2['Income'].mean()
md = data2['Income'].median()
print("median:",md)
print("mean:",mn)

if  mn < md:
    print ("negatively skewed")
else:
    print ("positively skewed")
```

##### Since the income distribution is positively skewed, imputation using median was preferred.

```python
median_value = data2['Income'].median() #get the median
data2['Income']=data2['Income'].fillna(median_value) #fill NaN values

#store to Cleaned_data.csv
data2.to_csv("C:/Users/Ivy/Desktop/data scientist/Python/Week 4/part2_cleaned_data.csv",index = False)

data2
```
